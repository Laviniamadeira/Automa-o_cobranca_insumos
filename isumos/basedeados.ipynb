```python
"""
==============================================================================
GERADOR DE PLANILHA FICTÍCIA - Exemplo de Portfólio
==============================================================================
Autora: Lavínia Madeira

O QUE ESSE SCRIPT FAZ:
Cria, do ZERO, uma planilha .xlsx com dados 100% fictícios, seguindo a
mesma estrutura de colunas usada no projeto "Robô de Cobrança de Insumos"
(Google Apps Script). Não usa nenhum dado real da Loggi — tudo é gerado
aleatoriamente. O resultado é seguro para você usar em prints, GitHub ou
como planilha de teste do seu script de Apps Script.

COMO USAR:
1. Numa célula do notebook, instale as dependências (uma vez só):
   !pip install openpyxl faker

2. Rode este script:
   python gerar_planilha_ficticia.py
   (ou, no Jupyter: cole o conteúdo numa célula e rode)

3. Vai gerar o arquivo "planilha_portfolio_exemplo.xlsx" na mesma pasta.
==============================================================================
"""

import random
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment
from faker import Faker

# ==============================================================================
# ⚙️ CONFIGURAÇÕES
# ==============================================================================

ARQUIVO_SAIDA = "planilha_portfolio_exemplo.xlsx"
NOME_ABA = "Disparo E-mails"
QUANTIDADE_PARCEIROS = 12  # quantas linhas fictícias gerar

PRECO_UNIT_SACA = 3.90
PRECO_UNIT_GAYLORD = 48.66

PERIODO_APURACAO = "01/06/2026 a 30/06/2026"
DATA_LIMITE = "15/07/2026"

# Linha onde os dados começam (segue o padrão do projeto original)
LINHA_INICIO_DADOS = 4

# ==============================================================================
# 🎲 GERADOR DE DADOS FICTÍCIOS
# ==============================================================================

fake = Faker("pt_BR")
random.seed(7)  # remova esta linha se quiser dados diferentes a cada execução


def gerar_nome_empresa(indice):
    return f"Empresa Fictícia {indice:02d}"


def gerar_email(indice):
    return f"contato@empresaficticia{indice:02d}.com"


def gerar_par_enviado_devolvido(maximo=120):
    enviado = random.randint(20, maximo)
    devolvido = random.randint(0, enviado)
    pendente = enviado - devolvido
    return enviado, devolvido, pendente


# ==============================================================================
# 🚀 MONTAGEM DA PLANILHA
# ==============================================================================

def gerar_planilha():
    wb = Workbook()
    aba = wb.active
    aba.title = NOME_ABA

    # --- Cabeçalho de controle (linhas 1 a 3, imitando o layout original) ---
    aba["A1"] = "Período de Apuração:"
    aba["B1"] = PERIODO_APURACAO
    aba["D1"] = "Data Limite:"
    aba["E1"] = DATA_LIMITE

    # --- Cabeçalho das colunas (linha 3) ---
    cabecalhos = {
        "A": "Nº", "B": "Nome do Parceiro (Leve)", "C": "",
        "D": "Sacas Enviadas", "E": "Sacas Devolvidas", "F": "Sacas Pendentes",
        "G": "Custo Sacas (R$)", "H": "Gaylords Enviados", "I": "Gaylords Devolvidos",
        "J": "Gaylords Pendentes", "K": "Custo Gaylords (R$)", "L": "Total Pendente (R$)",
        "M": "E-mail Financeiro", "N": "Status"
    }
    for coluna, titulo in cabecalhos.items():
        celula = aba[f"{coluna}3"]
        celula.value = titulo
        celula.font = Font(bold=True, color="FFFFFF")
        celula.fill = PatternFill(start_color="00AFFF", end_color="00AFFF", fill_type="solid")
        celula.alignment = Alignment(horizontal="center")

    # --- Linhas de dados fictícios ---
    for i in range(QUANTIDADE_PARCEIROS):
        linha = LINHA_INICIO_DADOS + i
        indice = i + 1

        sacas_env, sacas_dev, sacas_pen = gerar_par_enviado_devolvido(maximo=150)
        gay_env, gay_dev, gay_pen = gerar_par_enviado_devolvido(maximo=40)

        valor_sac = round(sacas_pen * PRECO_UNIT_SACA, 2)
        valor_gay = round(gay_pen * PRECO_UNIT_GAYLORD, 2)
        total_pendente = round(valor_sac + valor_gay, 2)

        aba.cell(row=linha, column=1, value=indice)
        aba.cell(row=linha, column=2, value=gerar_nome_empresa(indice))
        aba.cell(row=linha, column=4, value=sacas_env)
        aba.cell(row=linha, column=5, value=sacas_dev)
        aba.cell(row=linha, column=6, value=sacas_pen)
        aba.cell(row=linha, column=7, value=valor_sac)
        aba.cell(row=linha, column=8, value=gay_env)
        aba.cell(row=linha, column=9, value=gay_dev)
        aba.cell(row=linha, column=10, value=gay_pen)
        aba.cell(row=linha, column=11, value=valor_gay)
        aba.cell(row=linha, column=12, value=total_pendente)
        aba.cell(row=linha, column=13, value=gerar_email(indice))
        aba.cell(row=linha, column=14, value="Pendente")

    # --- Ajuste de largura das colunas para ficar legível no print ---
    larguras = {"A": 5, "B": 24, "C": 3, "D": 14, "E": 15, "F": 15,
                "G": 15, "H": 16, "I": 17, "J": 16, "K": 16, "L": 17,
                "M": 30, "N": 16}
    for coluna, largura in larguras.items():
        aba.column_dimensions[coluna].width = largura

    wb.save(ARQUIVO_SAIDA)
    print(f"Pronto! Planilha fictícia criada: '{ARQUIVO_SAIDA}'")
    print(f"{QUANTIDADE_PARCEIROS} parceiros fictícios gerados, com dados coerentes.")
    print("Pode usar essa planilha à vontade em prints, GitHub e para testar o script de Apps Script.")


if __name__ == "__main__":
    gerar_planilha()
```
